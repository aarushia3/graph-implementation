```python
# randomly generate a graph
nxG = nx.graph_atlas(150)
nx.draw(nxG)
plt.show()
```

```python
# dense_gnm_random_graph(n, m[, seed])
import random
n = random.randint(1, 10)
m = random.randint(1, 50)
seed = random.randint(1, 1000000)
nxG = nx.dense_gnm_random_graph(n, m, seed)
nx.draw(nxG)
plt.show()
```

```python
A = nx.adjacency_matrix(nxG)
A = A.todense()
print(A)
type(A)
```

```python
f = open("test\\test_graphs\\graph2.txt", "w")
f.write(str(A.shape[0]) + "\n")
for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        f.write(str(A[i][j]) + " ")
    f.write("\n")
f.close()
```
